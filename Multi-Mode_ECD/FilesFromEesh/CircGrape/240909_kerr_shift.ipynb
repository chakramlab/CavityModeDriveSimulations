```python
args = {'kerr': -0.0001 * 2 * np.pi,  #MHz
        'n_max': 30,
        'omega': 0.01 * 2 * np.pi,  #MHz
        }
```

```python
def P_0(alpha, t , args):
    return np.abs(np.exp(-np.abs(alpha)**2) *
                    np.sum([np.abs(alpha)**(2*n)/math.factorial(n) * np.exp(-1j * (-1 * args['kerr']/2* n**2 + args['omega'] *n) * t) for n in range(0, args['n_max'])])
                    )**2
```

```python
def P_0_new(alpha, t , args):
    return np.abs(np.exp(-np.abs(alpha)**2) *
                    np.sum([alpha**(2*n)/math.factorial(n) * np.exp(-1j * (-1 * args['kerr']/2* n**2 + args['omega'] *n) * t) for n in range(0, args['n_max'])])
                    )**2
```
